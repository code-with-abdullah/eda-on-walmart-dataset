# Preprocessing of Dataset

## Extract highest performing store from data

```python
preprocessed_walmart_data = walmart_data.copy()
```

```python
branch_df = preprocessed_walmart_data.groupby(["City", "Branch"]).agg({"Total":"sum"})
branch_df
```

```python
branch_df = branch_df.reset_index()
```

```python
branch_df.sort_values("Total")
```

```python
# Extract data of branch 'C' for Machine Learning

preprocessed_walmart_data = preprocessed_walmart_data[preprocessed_walmart_data["Branch"] == 'C']
```

## Drop unwanted columns from dataset

```python
preprocessed_walmart_data = preprocessed_walmart_data.drop(columns = ["Invoice ID"])
```

```python
preprocessed_walmart_data.head()
```

```python
preprocessed_walmart_data.head()
```

```python
preprocessed_walmart_data.describe()
```

```python
# Drop city and Branch as they are not required for ML model.
# All data rows will have the same city and branch so dropping these columns
```

```python
preprocessed_walmart_data = preprocessed_walmart_data.drop(columns = ["Branch", "City"])
```

```python
preprocessed_walmart_data.head()
```

```python
# Unit Price, Quantity, Date, Time are not Gross Margin %age, Gross Income are not required for ML model.
preprocessed_walmart_data = preprocessed_walmart_data.drop(columns = ["Unit price", 
                                                                      "Quantity",
                                                                      "Date",
                                                                      "Time",
                                                                      "gross margin percentage",
                                                                      "gross income"])
```

```python
#preprocessed_walmart_data = preprocessed_walmart_data.drop(columns= ["Index"])
```

```python
preprocessed_walmart_data.head()
```

```python
preprocessed_walmart_data = preprocessed_walmart_data.reset_index(drop=True)
```

```python
preprocessed_walmart_data.head()
```

```python
# As cogs + tax = total 
# We do not need these columns

preprocessed_walmart_data = preprocessed_walmart_data.drop(columns = ["Tax 5%", "cogs"])
```

```python
preprocessed_walmart_data.head()
```

```python
import sklearn
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, MinMaxScaler
```

# Normalization of Features

```python
X = preprocessed_walmart_data[["Customer type", "Gender", "Product line", "Payment", "Rating"]]
```

```python
y = preprocessed_walmart_data[["Total"]]
```

```python
X.head()
```

```python
y.head()
```

### Encoding Customer Type

```python
customer_type_le = LabelEncoder()
```

```python
X["Customer type"] = customer_type_le.fit_transform(X["Customer type"])
```

### Encoding Gender

```python
gender_le = LabelEncoder()
```

```python
X["Gender"] = gender_le.fit_transform(X["Gender"])
```

### One Hot Encoding Product Line

```python
product_line_le = LabelEncoder()
```

```python
X["Product line"] = product_line_le.fit_transform(X["Product line"])
```

```python
X["Product line"]
```

```python
product_line_ohe = OneHotEncoder(sparse=False)
```

```python
product_line = product_line_ohe.fit_transform(np.array(X["Product line"]).reshape(328, 1))
```

```python
# Removing dummy variable after One Hot Encoding
product_line = product_line[:, :-1]
```

```python
X = X.drop(columns = ["Product line"])
```

```python
product_line
```

```python
X
```

```python
product_line = product_line.reshape(5, 328)
```

```python
product_line.shape
```

```python
X["product line 0"] = product_line[0]
X["product line 1"] = product_line[1]
X["product line 2"] = product_line[2]
X["product line 3"] = product_line[3]
X["product line 4"] = product_line[4]
```

```python
X
```

### One Hot Encoding Payment

```python
payment_le = LabelEncoder()
X["Payment"] = payment_le.fit_transform(X["Payment"])
```

```python
payment_ohe = OneHotEncoder(sparse=False)
```

```python
payment = payment_ohe.fit_transform(np.array(X["Payment"]).reshape(328, 1))
```

```python
payment = payment[:, :-1]
```

```python
payment = payment.reshape(2, 328)
```

```python
X = X.drop(columns = ["Payment"])
```

```python
X["payment 0"] = payment[0]
X["payment 1"] = payment[1]
```

```python
X
```

```python
rating_scaler = MinMaxScaler()
```

```python
X["Rating"] = rating_scaler.fit_transform(np.array(X["Rating"]).reshape(328, 1))
```

```python
X
```

```python
X.to_csv("preprocessed_dependant_variables.csv")
```
